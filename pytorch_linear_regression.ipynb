```python
import torch
from torch.autograd import Variable
```

```python
x_data = Variable(torch.tensor([[1.0],[2.0],[3.0]]))
y_data = Variable(torch.tensor([[2.0],[4.0],[6.0]]))
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        #in this constructor we instantiate two linear nn module
        super(Model,self).__init__()
        self.linear = torch.nn.Linear(1,1)#one inpuyt one output
    def forward(self,x):
        y_pred = self.linear(x)
        return y_pred
model = Model()
```

```python
"""
    construct our loss funtion and our optimizers.
    The call to mmodel.parameters() in the SGD constructor will contain
    the learnable parameters of the two nn.Linear modules which are
    the members of the module.
"""
criterion = torch.nn.MSELoss(size_average = False)
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
#training loop
for epoch in range(500):
    #forward pass; compute preddicted y by passing x to the model
    y_pred = model(x_data)
    
    #compute and print loss
    loss = criterion(y_pred,y_data)
    print(epoch,loss.data[0])
    
    #zero gradients, perform a backward pass and update the weights
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
#after training
hour_var = Variable(torch.Tensor([4.0]))
print ("predictedd value for input 4 is:",model(hour_var).data[0][0])
```
